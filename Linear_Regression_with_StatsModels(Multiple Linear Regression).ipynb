# Multiple Linear Regression


```python
import pandas as pd
```

```python
df_boston = pd.read_csv('Boston House Prices.csv')
```

```python
df_boston
```

```python
import statsmodels.api as sm
```

```python
y = df_boston['Value']
X = df_boston[['Rooms','Distance']]
```

```python
X = sm.add_constant(X)
```

```python
lm = sm.OLS(y,X).fit()
```

```python
lm.summary()
```

# Linear Regression with Scikit-learn

```python
from sklearn import linear_model
```

```python
y = df_boston['Value']
X = df_boston[['Rooms','Distance']]
```

```python
lm = linear_model.LinearRegression()
lm.fit(X,y)
```

```python
lm.predict(X)[:5]
```

```python
# r2 score
lm.score(X,y)
```

```python
#coefficents
lm.coef_
```

```python
#intercept
lm.intercept_
```
